## Invoking the SEVIR Nowcast API using FastAPI web-framework

### Test case 1:

```python
import requests

# Define parameters
params_test1 = {
 "llcrnrlat":37.318363,
 "llcrnrlon":-84.224203, 
 "urcrnrlat":40.087573 ,
 "urcrnrlon":-79.052692,
 "time_utc":"2019-06-02 18:33:00"
}
```

```python
# Post request
nowcast_test1 = requests.post(f"http://127.0.0.1:8000/nowcast/", json = params_test1)

# JSON output
sevir_output_test1 = nowcast_test1.json()
sevir_output_test1.keys()
```

```python
# H5 file output path
print("Path for nowcasted output in h5 file:\n", sevir_output_test1['nowcast_path'])

# GIF file output path
print("Path for nowcasted output in GIF format:\n", sevir_output_test1['gif_path'])
```

```python
# GIF display
from IPython.display import Image

Image(open(sevir_output_test1['gif_path'], 'rb').read())
```
